```python
def format_strings(string):
  return string.replace("['", "").replace("']", "").replace('["', "")\
        .replace('"]', "").replace('", "', ". ").replace("', '", ". ")\
        .replace("\", '", " ").replace("', \"", " ").replace('"', "")\
        .replace("\\", "")

items['description'] = items['description'].apply(format_strings)
items['description'].iloc[3]
```

```python
items['feature'].iloc[0]
```

```python
items['feature'] = items['feature'].apply(format_strings)
items['feature'].iloc[0]
```

```python
items['description'] = items['description'] + ' ' + items['feature']
items.drop('feature', axis=1, inplace=True)
items['description'].iloc[0]
```
