```python
import pandas as pd
import gzip
import json
```

```python
from google.colab import  drive

drive.mount('/drive')

ratings = pd.read_csv('/drive/My Drive/Datasets/amazon_fashion/clothing_shoes_jewelry_ratings_cleaned.csv')
ratings
```

```python
unique_items = ratings['asin'].unique()
unique_items
```

```python
from google.colab import output


META_DATA_LEN = 2685059


data = []
with gzip.open('/content/meta_Clothing_Shoes_and_Jewelry.json.gz') as f:
  i = 0
  for l in f:
    i += 1
    item = json.loads(l.strip())
    if item['asin'] in unique_items:
      data.append(item)
    
    if i % 100 == 0:
      output.clear()
      print(f"{round(i/META_DATA_LEN*100,2)}%")
data
```

```python
df = pd.DataFrame.from_dict(data)
df.head()
```

```python
df.to_csv('/drive/My Drive/Datasets/amazon_fashion/clothing_shoes_jewelry_meta_data.csv', index=False)
```
