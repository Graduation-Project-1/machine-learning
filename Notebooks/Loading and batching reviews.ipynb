```python
import pandas as pd
import gzip
import json
```

```python
from google.colab import  drive

drive.mount('/drive')

ratings = pd.read_csv('/drive/My Drive/Datasets/clothing_dataset/clothing_shoes_jewelry_ratings_cleaned_final.csv')
ratings
```

```python
unique_reviewers = ratings['reviewerID'].unique()
unique_reviewers
```

```python
from google.colab import output


META_DATA_LEN = 32292099


data = []
with gzip.open('/content/Clothing_Shoes_and_Jewelry.json.gz') as f:
  i = 0
  for l in f:
    i += 1
    review = json.loads(l.strip())
    if review['reviewerID'] in unique_reviewers:
      data.append(review)
    
    if i % 1000 == 0:
      output.clear()
      print(f"{round(i/META_DATA_LEN*100,2)}%")
data
```

```python
df = pd.DataFrame.from_dict(data)
df.head()
```

```python
unique_items = ratings['asin'].unique()
unique_items
```

```python
reviews = df[df['asin'].isin(unique_items)]
reviews
```

```python
reviews.drop(['image', 'vote', 'summary', 'verified'], axis=1, inplace=True)
reviews
```

```python
styles = list(reviews['style'])
reviews.drop('style', axis=1, inplace=True)
styles[:5]
```

```python
styles = pd.json_normalize(styles)
styles
```

```python
col_names = {'Size:' : 'size',
             'Color:' : 'color',
             'Width:' : 'wdith',
             'Format:' : 'format',
             'Size Name:' : 'sizeName',
              'Style:' : 'style'
              }
              
styles.rename(columns = col_names, inplace = True)
styles.set_index(reviews.index, inplace = True)
styles
```

```python
reviews = pd.concat([reviews, styles], axis=1)
reviews
```

```python
reviews.duplicated().sum()
```

```python
reviews.drop_duplicates(inplace=True)
reviews.duplicated().sum()
```

```python
reviews.isna().sum()
```

```python
reviews.drop(['color', 'wdith', 'size', 'style', 'format'], axis=1, inplace=True)
reviews = reviews.dropna()
reviews
```

```python
reviews.isna().sum()
```

```python
reviews.to_csv('/drive/My Drive/Datasets/clothing_dataset/clothing_shoes_jewelry_reviews.csv', index=False)
```
