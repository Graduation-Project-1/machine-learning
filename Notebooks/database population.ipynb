```python
items = items.replace({'brand': db_brands_dict})
items.head()
```

```python
items['numberOfReviews'] = 0
items['averageRate'] = 0
items['numberOfLikes'] = 0
items['gender'] = 'male'

items.head()
```

```python
def fromat_price(string):
  string = string.replace('$', '').replace(',', '')
  dash_pos = string.find(' - ')   
  if(dash_pos != -1):
    price_list = string.split(' - ')
    price = (float(price_list[0]) + float(price_list[1]))/2
  else:
    price = float(string)

  return int(price * 24.75)

items['price'] = items['price'].apply(fromat_price)
items['price'].head(10)
```

```python
items_dict = items.to_dict('records')
items_dict[0]
```

```python
def format_categories(string):
  categories_list = string.replace("[", "").replace("]", "").replace("'", "")\
                    .replace('"', "").split(", ")
  
  id_categories_list = []
  for category in categories_list:
    id_categories_list.append(db_categories_dict[category])
  
  return id_categories_list

for item in items_dict:
  item['categoryList'] = format_categories(item['categoryList'])

items_dict[0]
```

```python
def format_images(string):
  images = string.replace("[", "").replace("]", "").replace("'", "")\
            .replace('"', "").split(", ")
  
  return images


for item in items_dict:
  item['images'] = format_images(item['images'])
  item['sizes'] = []
  item['colors'] = []

items_dict[:2]
```
